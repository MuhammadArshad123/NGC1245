```python
#Dependencies
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def displayData(spectral_count):
  keys = spectral_count.keys()
  values = spectral_count.values()
  fig = plt.figure(figsize = (10, 5))
  plt.bar(keys, values, color ='blue',
        width = 0.4)
  plt.xlabel("Spectral Types")
  plt.ylabel("Count")
  plt.title("Amount of Stars in NGC 1245 Classified by Spectral Type")
  plt.show()
```

```python
SPECTRAL_TYPES = {1.6:'M',1.0:'K',0.7:'G',0.5:'F',0.2:'A',-0.2:'B',-0.4:'O'}
spectral_count = {'M':0,'K':0,'G':0,'F':0,'A':0,'B':0,'O':0,'[UNK]':0}
star_spectral_types = []

#Reading the data from the csv file and parsing the column needed
star_data = pd.read_csv('ngc1245.csv')
color_indexes = star_data['B-V_corrected'].fillna(0)

#Iterates over the star's color indexes and classifies them accordingly
for i in color_indexes:
  if i == 0:
    star_spectral_types.append('[UNK]')
    spectral_count['[UNK]']+=1
  else:
    found = False
    for j in SPECTRAL_TYPES.keys():
      if i >= j:
        star_spectral_types.append(SPECTRAL_TYPES[j])
        spectral_count[SPECTRAL_TYPES[j]]+=1
        found = True
        break
    if found == False:
      star_spectral_types.append('O')

for i in spectral_count.keys():
  print('Spectral Type: '+i+', Amount: '+str(spectral_count[i]))
displayData(spectral_count)




```
